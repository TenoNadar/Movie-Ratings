```python
df_by_genre=data.loc[:,"CVotes10":"VotesnUS"]
df_by_genre[["genre_1","genre_2","genre_3"]]=data[["genre_1","genre_2","genre_3"]]
df_by_genre.head()
```

```python
df_by_genre["cnt"]=1
```

```python
df_by_genre.head()
```

```python
df_grp1=df_by_genre.groupby("genre_1").aggregate(np.sum)
df_grp2=df_by_genre.groupby("genre_2").aggregate(np.sum)
df_grp3=df_by_genre.groupby("genre_3").aggregate(np.sum)
```

```python
df_add=df_grp1.add(df_grp2,fill_value=0)
```

```python
df_add=df_add.add(df_grp3,fill_value=0)
```

```python
df_add.head()
```

```python
genre_top_10=df_add.loc[df_add["cnt"]>10]
```

```python
genre_top_10.head()
```

```python
genre_top_10.head()
```

```python
genre_top_10.iloc[:,0:-1]=genre_top_10.iloc[:,0:-1].divide(genre_top_10["cnt"],axis=0)
genre_top_10
```

```python
genre_top_10.loc[:,"VotesM":"VotesnUS"]=round(genre_top_10.loc[:,"VotesM":"VotesnUS"],2)
```

```python
genre_top_10
```

```python
genre_top_10[genre_top_10.loc[:,"CVotes10":"CVotesnUS"].columns]=genre_top_10[genre_top_10.loc[:,"CVotes10":"CVotesnUS"].columns].astype(int)
```
